## Creating a Noise Column

```python
def noise_column(x):
    if x.startswith('Noise'):
        return x

complaints['Noise'] = complaints['Complaint Type'].apply(noise_column)
```

```python
complaints.Noise.value_counts()
```

```python
noise_complaint_df = complaints[is_noise][['Complaint Type', 'Borough', 'Created Date', 'Descriptor']]
```

## Which Borough has the highest noise complaint considering all sources of noise

```python
noise_complaint_df['Borough'].value_counts()
```

```python
noise_complaint_counts = noise_complaint_df['Borough'].value_counts()
complaint_counts = complaints['Borough'].value_counts()
```

```python
(noise_complaint_counts / complaint_counts.astype(float)).plot(kind='bar')
```
